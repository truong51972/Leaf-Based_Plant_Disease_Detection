```python
data = {'Hàng 3': {'Bệnh bạc lá sớm': 1, 'Cây tốt': 5}, 'Hàng 1': {'Bệnh tàn rụi muộn': 1, 'Cây tốt': 5}, 'Hàng 4': {'Virus TYLCV (Tomato yellow leaf curl virus)': 1, 'Bệnh đốm vòng hại khoai tây': 1, 'Cây tốt': 5}, 'Hàng 2': {'Virus khảm cà chua ToMV': 2, 'Cây tốt': 5}}
data.keys()
```

```python
categories = []  # List to store categories (x-axis labels)
counts = []  # List to store counts (y-axis values)

for row in data.keys():  # Iterate through each row (key) in the dictionary
    for disease in data[row]:  # Iterate through diseases (values) for each row
        categories.append(f"{row} - {disease}")  # Create combined category labels
        counts.append(data[row][disease])  # Append corresponding count
```

```python
plt.figure(figsize=(8, 6))  # Set figure size
plt.bar(categories, counts, color='skyblue')  # Create bar plot with skyblue color
plt.xlabel('Categories (Row - Disease)')  # Set x-axis label
plt.ylabel('Count')  # Set y-axis label
plt.title('Disease Distribution by Row')  # Set plot title
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better presentation
```

```python
row_totals = {row: sum(counts.values()) for row, counts in data.items()}

# Extract row names and total counts
rows = list(row_totals.keys())
totals = list(row_totals.values())

# Create the bar data
plt.figure(figsize=(8, 6))
plt.bar(rows, totals, color='skyblue')
plt.xlabel('Row')
plt.ylabel('Total Disease Count')
plt.title('Total Disease Count by Row')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Create a list of unique disease names
diseases = list(set([disease for row_data in data.values() for disease in row_data.keys()]))

# Create a matrix with zeros for each row and disease
data_matrix = np.zeros((len(data), len(diseases)))

# Fill the matrix with disease counts
for i, row in enumerate(data):
    for j, disease in enumerate(diseases):
        data_matrix[i, j] = data[row].get(disease, 0)

# Create a colormap
colormap = plt.cm.get_cmap('tab10', len(diseases))

# Create the stacked bar data
fig, ax = plt.subdatas()
bottom = np.zeros(len(data))
for i, disease in enumerate(diseases):
    ax.bar(range(len(data)), data_matrix[:, i], bottom=bottom, label=disease, color=colormap(i))
    bottom += data_matrix[:, i]

ax.set_xticks(range(len(data)))
ax.set_xticklabels(data.keys(), rotation=45, ha='right')
ax.set_ylabel('Count')
ax.set_title('Disease Distribution by Row')
ax.legend()
plt.tight_layout()
plt.show()
```
